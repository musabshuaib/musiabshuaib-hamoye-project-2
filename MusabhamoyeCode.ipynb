```python
# Define features and target
X = df.drop(columns="Appliances")
y = df["Appliances"]
```

```python
from sklearn.model_selection import train_test_split

# Train-test split (70-30)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.preprocessing import MinMaxScaler

# Create an instance of the MinMax scaler
scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
lr = LinearRegression()
lr.fit(X_train_scaled, y_train)
```

```python
train_predictions = lr.predict(X_train_scaled)
```

```python
from sklearn.metrics import mean_absolute_error

# Calculating the Mean Absolute Error 
mae = mean_absolute_error(y_train, train_predictions)
print("Mean Absolute Error (Training Set):", mae)
```

```python
#Calculating the RMSE for the train data 
rmse_train = np.sqrt(mean_squared_error(y_train, train_predictions))
print("Training RMSE:", rmse_train)
```

```python
from sklearn.metrics import mean_absolute_error

# Prediction on the test set
test_predictions = lr.predict(X_test_scaled)
```

```python
# Calculating the Mean Absolute Error for test data
mae_test = mean_absolute_error(y_test, test_predictions)

print("Mean Absolute Error for the test set:", mae_test)
```

```python
# Calculating the Root mean squared error for test data to 3 D.P
rmse_test = np.sqrt(mean_squared_error(y_test, test_predictions))
print("Test RMSE:", rmse_test)
```

```python
#checking for overfitting
from sklearn.metrics import r2_score

r2_train = r2_score(y_train, train_predictions)
r2_test = r2_score(y_test, test_predictions)

print("Train RÂ² Score:", r2_train)
print("Test RÂ² Score:", r2_test)
```

```python
#The RÂ² scores above shows that the Model is Under fitted
```

```python
from sklearn.linear_model import Ridge

ridge = Ridge(alpha=1.0)  # default alpha=1.0
ridge.fit(X_train, y_train)

y_pred_test = ridge.predict(X_test)

rmse_test = np.sqrt(mean_squared_error(y_test, y_pred_test))
print("Test RMSE (Ridge):", rmse_test)
```

```python
from sklearn.linear_model import Lasso

lasso = Lasso(alpha=1.0)  
lasso.fit(X_train, y_train)

# Getting the coefficients (feature weights)
coefficients = lasso.coef_

# Checking the amount of non-zero weights
non_zero = sum(coeff != 0 for coeff in coefficients)

print("Number of features with non-zero weights:", non_zero)
```

```python
# Predictions on test set with lasso
y_pred_test = lasso.predict(X_test)

# Calculating the RMSE
rmse_test = np.sqrt(mean_squared_error(y_test, y_pred_test))
print("Test RMSE (Lasso):", rmse_test)
```
